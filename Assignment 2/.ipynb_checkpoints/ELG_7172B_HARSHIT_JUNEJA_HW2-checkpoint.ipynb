To make a bag of words, we will use the code in [1]

```python
#Importing libraries
#nltk for processing the messages
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
#for calculating frequencies of tokens (words)
from collections import Counter
#matlplotlib and wordcloud for visualization
import matplotlib.pyplot as plt
from wordcloud import WordCloud
#pandas is used to import data
import pandas as pd
#numpy is used to split the dataset into train and test sets
import numpy as np
#log is used in predictions to make calculations easier
from math import log
```

The first step is to load the data from the CSV file [2 - paper] [3 - webpage]

```python
msgs = pd.read_csv('SMSSpamCollection.tsv', sep = '\t', header = None, names=['label','message'], quoting = 3, encoding = 'latin-1')
print("\nSize of input = ", msgs.shape)
print(msgs['label'].value_counts())
print("\nSample data values \n\n", msgs.head())
```

Now, we split the data into two classes: training and test

```python
prng = np.random.RandomState(42) #seen the prng for repeatable results
split = 0.75 #the proportion of samples assigned to training set
totalMsgs = msgs['message'].shape[0]
trainIndex, testIndex = list(), list()
for i in range(msgs.shape[0]):
    if prng.uniform(0,1) < split:
        trainIndex += [i]
    else:
        testIndex += [i]
trainData = msgs.loc[trainIndex]
testData = msgs.loc[testIndex]
trainData.reset_index(inplace = True)
trainData.drop(['index'], axis = 1, inplace = True)
print(trainData['label'].value_counts())
print("\nSample values from training set \n\n", trainData.head())
```

```python
testData.reset_index(inplace = True)
testData.drop(['index'], axis = 1, inplace = True)
print(testData['label'].value_counts())
print("\nSample values from testing set \n\n", testData.head())
```

<br>Next, we write a small function to process a message and extract tokens

```python
def process_message(message, lower_case = True, stem = False, stop_words = True, gram = 1):
    if lower_case: #we convert the message to lower case
        message = message.lower()
    #the message is tokenized into words, all punctuation characters are neglected
    words = word_tokenize(message)
    #ignore all tokens of length 2 or less
    words = [w for w in words if len(w) > 2]
    #Porter stemming is used to reduce words to their root form, e.g. 'go' and 'going' indicate the same activity
    if stem:
        stemmer = PorterStemmer()
        words = [stemmer.stem(word) for word in words]
    #stop words are very common words in a language which do not provide much information as to the nature of a message
    if stop_words:
        sw = stopwords.words('english')
        words = [word for word in words if word not in sw]
    #n-gram is a token containing n words, these are used to improve accuracy
    #e.g. 'good' and 'not good' are opposite in meaning, it is better to consider 'not good' as one token
    if gram > 1:
        w = []
        for i in range(len(words) - gram + 1):
            w += [' '.join(words[i:i + gram])]
        words = w
    return words
```

For a word w, the probability of word is given by $P(w)$,<br><br>
\begin{align}
P(w) &= \frac{\text{Total number of occurences of w in dataset}}{\text{Total number of words in dataset}} \tag{1}\\
\end{align}
<br>and, the probability of a word w given that a message is spam is given by<br><br>
\begin{align}
P(w|spam) &= \frac{\text{Total number of occurences of w in spam messages}}{\text{Total number of words in spam messages}} \tag{2}\\
\end{align}
<br><br>
However, if we encounter a new word in our test data set, which is not present in the training dataset, then $P(w) = 0$, $P(w|spam) = 0$, and $P(w|ham) = 0$ which would lead to a breakdown of our classification model for that particular element. Therefore, we shall use additive smoothing for prediction posterior probability:<br><br>
\begin{align}
\mathbb{P}(w|spam) &= \frac{\text{Total number of occurences of w in spam messages} + \alpha}{\text{Total number of words in spam messages} + \alpha\text{(Total number of words in dataset)}} \tag{3}\\
\end{align}
<br><br>
Similarly for predicting ham messages,
\begin{align}
\mathbb{P}(w|ham) &= \frac{\text{Total number of occurences of w in ham messages} + \alpha}{\text{Total number of words in ham messages} + \alpha\text{(Total number of words in dataset)}} \tag{4}\\
\end{align}
<br><br>
Now, we will process the training data set.<br>
We will only calculate frequencies, and use these frequencies in calculating probabilites to make class predictions on the test set.<br>

```python
def process_dataset(Dataset):
    words = []
    words_spam = []
    words_ham =[]
    spam_count = 0
    ham_count = 0
    for i in range (Dataset.shape[0]):
        words += process_message(Dataset.iloc[i]['message'])
        if Dataset.iloc[i]['label'] == 'spam':
            words_spam += process_message(Dataset.iloc[i]['message'])
            spam_count = spam_count + 1
        elif Dataset.iloc[i]['label'] == 'ham':
            words_ham += process_message(Dataset.iloc[i]['message'])
            ham_count = ham_count + 1
    words_freq = dict(Counter(words))
    words_ham_freq = dict(Counter(words_ham))
    words_spam_freq = dict(Counter(words_spam))
    spam_prob = spam_count/(i+1)
    ham_prob = ham_count/(i+1)
    return ham_prob, spam_prob, words_freq, words_ham_freq, words_spam_freq
```

Now, let us invoke the function on the training set and get the frequencies:<br>

```python
p_ham, p_spam, freq_words, freq_words_ham, freq_words_spam = process_dataset(trainData)
```

Now, we will create a predict function to predict whether a new message is spam or ham.<br>
We will use Naive Bayes classifier:<br>
\begin{align}
P(Spam|w1,w2,...,wn) \propto \frac{P(spam)*P(w1,w2,...,wn|spam)}{P(w1,w2,...,wn)} \tag{5}\\
\end{align}
<br>However, since our motive is to classify messages a spam or ham, we need not calculate the exact probability $P(Spam|w1,w2,...,wn)$. We can use:<br><br>
\begin{align}
P(Spam|w1,w2,...,wn) \propto P(spam)*\mathbb{P}(w1,w2,...,wn|spam) \tag{6}\\
\end{align}
<br>
\begin{align}
P(ham|w1,w2,...,wn) \propto P(ham)*\mathbb{P}(w1,w2,...,wn|ham) \tag{7}\\
\end{align}
<br>
We will classify a message as spam or ham based upon the larger value from equation 6 and equation 7. Here, since we are using Naive Bayes classifier,<br><br>
\begin{align}
\mathbb{P}(w1,w2,...,wn|spam) &= \mathbb{P}(w1|spam)*\mathbb{P}(w2|spam)*...*\mathbb{P}(wn|spam) \tag{8}\\
\end{align}
<br><br>
Where $\mathbb{P}(wn|spam)$ is defined in equation 3. The equations for ham are similar.
Further, in the actual code below, instead of performing multiplication directly, we will take logarithms for they are easier to compute. This does not affect the final answer.<br>
The only limitation with using logarithms is that alpha parameter <b><u>cannot</u></b> be zero

```python
def predict(processed_message, prob_ham, prob_spam, f_words, f_words_ham, f_words_spam, alpha):
    n = len(f_words)
    n_ham = len(f_words_ham)
    n_spam = len(f_words_spam)
    pm_words_freq = dict(Counter(processed_message))
    prob_pm_words_ham = {}
    res_h = 1
    for k in pm_words_freq:
        pm_words_freq[k] = f_words_ham.get(k,0) + alpha
        prob_pm_words_ham[k] = log(pm_words_freq[k]/(n_ham + alpha * n))
        #res_h = res_h * prob_pm_words_ham[k]
    prediction_ham = log(prob_ham)+sum(prob_pm_words_ham.values())#* res_h
    prob_pm_words_spam = {}
    res_sp = 1
    for k in pm_words_freq:
        pm_words_freq[k] = f_words_spam.get(k,0) + alpha
        prob_pm_words_spam[k] = log(pm_words_freq[k]/(n_spam + alpha * n))
        #res_sp = res_sp * prob_pm_words_spam[k]
    prediction_spam = log(prob_spam)+sum(prob_pm_words_spam.values()) #* res_sp
    if prediction_ham > prediction_spam:
        return 'ham'
    else:
        return 'spam'
```

```python
def bulk_predict(Dataset, alpha = 1):
    a = alpha
    predictions = Dataset.copy()
    predictions['predicted_label'] = "-"
    for i in range(Dataset.shape[0]):
        pm = process_message(Dataset.iloc[i]['message'])
        predictions.loc[i]['predicted_label'] = predict(pm, p_ham, p_spam, freq_words, freq_words_ham, freq_words_spam, a)
    return predictions
```

```python
naive_classifier_output = bulk_predict(testData, alpha = 1)
confusion_matrix = pd.crosstab(naive_classifier_output['label'], naive_classifier_output['predicted_label'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

From the confusion matrix we see that most of the ham messages were labeled correctly. However, 30 out of 196 spam messages were incorrectly labeled as ham. Let us use two different values of alpha, 0.2 and 2, to see if the results change.

```python
naive_classifier_output = bulk_predict(testData, alpha = 0.2)
confusion_matrix = pd.crosstab(naive_classifier_output['label'], naive_classifier_output['predicted_label'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

```python
naive_classifier_output = bulk_predict(testData, alpha = 2)
confusion_matrix = pd.crosstab(naive_classifier_output['label'], naive_classifier_output['predicted_label'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

We observe that the results get better with a smaller value of alpha. However, there seems to be a trade-off here, as the mislabeling of ham increases with a smaller value of alpha. Let us try one more value, alpha = 0.01

Now, we will use the naive bayes classifier provided with the python library scikit-learn [4].

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
```

Since scikit-learn does <b>not</b> support features present in the test set but missing in the training set. We will have to build and use dictionary using the training dataset for the features matrix for both training and test sets. This is because Scikit-learn's implementation uses matrix multiplication internally and the dimensions must match!

```python
#parts of the code below have been adapted from [5]
def get_features(dataset, dictionary):
    features_matrix = np.zeros((dataset.shape[0],len(dictionary)))
    # i is the message j is the feature (token/word)
    for i in range(dataset.shape[0]):
        words = process_message(dataset.iloc[i]['message'])
        for word in words:
            wordID = 0
            for j, d in enumerate(dictionary):
                if d == word:
                    wordID = j
                    features_matrix[i,j] = words.count(word)
    return features_matrix
```

```python
training_set_features = get_features(trainData, freq_words)
train_labels = trainData['label'].to_numpy()
```

```python
scikit_naive_bayes = MultinomialNB(alpha = 1)
scikit_naive_bayes.fit(training_set_features, train_labels)
```

```python
test_set_features = get_features(testData, freq_words)
print(test_set_features.shape)
test_labels = testData['label'].to_numpy()
result = scikit_naive_bayes.predict(test_set_features)
```

```python
print(confusion_matrix(test_labels,result, labels=["ham", "spam"]))
```

We observe that the Naive Bayes classifier of scikit-learn gives performance comparable to the one we created from scratch. However, the scikit-classifier is limited as it cannot handle values present in the test set but missing from the training input. <br>
