<h5>Reading and pre-processing the data</h5>
We will take 40 labeled beats per patient for training, for a total of 160 heartbeats (80 normal, 80 paced). However, we will read more samples for all the beats may not have the required label. A list of labels for this dataset is given at [5]. The data is originally sampled at a frequency of 360 Hz, that is, 360 samples per second.<br>
First, we will plot some data just to see how the given ECG signal looks like.

```python
train_patients = ['101', '106', '102', '104']
test_patients = ['103', '105', '107', '217']
# we will take 40 samples per patient for training, given data is sampled at 360 Hz
# We need a minimum of 40*360 samples
# We will read 60*360 =21600 samples because not all samples may have the same label
sample_start = 0
sample_end = 21600
# Plot 2 seconds of ECG data for patient 101 with labels - regular heart beats
record = wf.rdrecord('data/101', channels = [0,1], sampfrom = sample_start, sampto = int(sample_end/30))
annotations = wf.rdann('data/101', 'atr', sampfrom = sample_start, sampto = int(sample_end/30), shift_samps = True)
wf.plot_wfdb(record = record, annotation = annotations, plot_sym = True, time_units = 'seconds',
             title='Record 101 from MIT-BIH dataset - regular heartbeats')
# Plot 2 seconds of ECG data for patient 102 with labels - paced heart beats
record = wf.rdrecord('data/102', channels = [0, 1], sampfrom = sample_start, sampto = int(sample_end/30))
annotations = wf.rdann('data/102', 'atr', sampfrom = sample_start, sampto = int(sample_end/30), shift_samps = True)
wf.plot_wfdb(record = record, annotation = annotations, plot_sym = True, time_units = 'seconds',
             title='Record 102 from MIT-BIH dataset - paced heartbeats')
```

We will only use the first channel to train the autoencoder. As the next step, we will create a fifth order Butterworth bandpass filter with a passband of 1 Hz to 60 Hz using scipy to remove noise from the ECG signal.

```python
# Bandpass filter to remove noise
fs = 360
nyq = 0.5*360
low_cutoff = 1
high_cutoff = 60
order = 5
B, A = sp.signal.butter(order, [low_cutoff/nyq, high_cutoff/nyq], btype='bandpass', analog=False, output='ba')
w, h = sp.signal.freqz(B, A)
plt.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
plt.plot(low_cutoff, 0.5*np.sqrt(2), 'ko')
plt.axvline(low_cutoff, color='k')
plt.plot(high_cutoff, 0.5*np.sqrt(2), 'ko')
plt.axvline(high_cutoff, color='k')
plt.xlim(-2, 0.5*fs)
plt.title("Bandpass Filter Frequency Response")
plt.xlabel('Frequency [Hz]')
plt.grid()
```

Let us test the bandpass filter with a noisy test signal - a sum of 33 Hz, 90 Hz and 150 Hz waveforms, and see if the Butterworth can extract the 33 Hz waveform.

```python
# Test the bandpass filter on a noisy test signal
T = 0.15   # seconds
n = int(T * fs) # total number of samples
t = np.linspace(0, T, n, endpoint=False)
# Noisy data.  We want to recover the 33 Hz signal from this.
data = 1.2*np.sin(33*2*np.pi*t) + 1.25*np.cos(90*2*np.pi*t) + 0.75*np.sin(150*2*np.pi*t)
# Filter the data, and plot both the original and filtered signals.
y = sp.signal.lfilter(B, A, data)
plt.plot(t, data, 'b-', label='noisy signal')
plt.plot(t, y, 'g-', linewidth=2, label='filtered signal')
plt.xlabel('Time [sec]')
plt.grid()
plt.legend(loc='best')
```

Now, we will read and preprocess the training data. We will perform the following steps:
* Read 60 seconds worth of data for each training patient
* Remove noise using the Bandpass filter
* Normalize the signal to range [-1,1]
* Extract 0.5 second waveforms as training examples, that is, 180 samples, with the label
* Downsample from 360 Hz to 60 Hz
* Label normal beat as 0, paced beat as 1

```python
#upper and lower bound for normalization
lower_bnd = -1
upper_bnd = 1
# Preprocessing training data
k = 0
training_data = np.zeros((160, 30))
training_labels = np.zeros((160, 1))
for pat in range(len(train_patients)):
    signal, fields = wf.rdsamp('data/'+train_patients[pat], sampfrom = sample_start, sampto = sample_end, channels = [0])
    filt_signal = sp.signal.lfilter(B, A, signal)
    filt_signal = wf.processing.normalize_bound(filt_signal, lb = lower_bnd, ub = upper_bnd)
    annotations = wf.rdann('data/'+train_patients[pat], 'atr', sampfrom = sample_start, sampto = sample_end, shift_samps = True)
    j = 0
    for i in range(len(annotations.sample)):
        if pat < 2 and annotations.symbol[i] == "N" and i > 1 and j < 40:
            marker = annotations.sample[i]
            l = max(0, marker - 90)
            h = l + 180
            train_signal = filt_signal[l:h]
            # downsampling from 360 samples per second to 60 samples per second
            train_signal = train_signal.reshape(30, 6)[:,0]
            training_data[k] = train_signal
            training_labels[k] = 0
            j += 1
            k += 1
        if pat >= 2 and annotations.symbol[i] == "/" and i > 1 and j < 40:
            marker = annotations.sample[i]
            l = max(0, marker - 90)
            h = l + 180
            train_signal = filt_signal[l:h]
            # downsampling from 360 samples per second to 60 samples per second
            train_signal = train_signal.reshape(30, 6)[:,0]
            training_data[k] = train_signal
            training_labels[k] = 1
            j += 1
            k += 1
```

Now, we will plot the data to see the alignment and similarity of waveforms:

```python
# Plot all Normal beats for patient 101
for i in range(0, 40):
    plt.plot(training_data[i])
_ = plt.title('All Normal heartbeats for patient 101')
```

```python
# Plot all Paced beats for patient 102
for i in range(80, 120):
    plt.plot(training_data[i])
_ = plt.title('All Paced heartbeats for patient 102')
```

Similary, we read and pre-process the test data with the only difference, that we only take 10 examples per patient, for a total of 40 test examples.

```python
# Preprocessing test data, for test set we will take 10 heartbeat samples per patient, for a total of 40 heartbeats
k = 0
test_data = np.zeros((40, 30))
test_labels = np.zeros((40, 1))
for pat in range(len(test_patients)):
    signal, fields = wf.rdsamp('data/'+test_patients[pat], sampfrom = sample_start, sampto = sample_end, channels = [0])
    filt_signal = sp.signal.lfilter(B, A, signal)
    filt_signal = wf.processing.normalize_bound(filt_signal, lb = lower_bnd, ub = upper_bnd)
    annotations = wf.rdann('data/'+test_patients[pat], 'atr', sampfrom = sample_start, sampto = sample_end, shift_samps = True)
    j = 0
    for i in range(len(annotations.sample)):
        if pat < 2 and annotations.symbol[i] == "N" and i > 1 and j < 10:
            marker = annotations.sample[i]
            l = max(0, marker - 90)
            h = l + 180
            test_signal = filt_signal[l:h]
            # downsampling from 360 samples per second to 60 samples per second
            test_signal = test_signal.reshape(30, 6)[:,0]
            test_data[k] = test_signal
            test_labels[k] = 0
            j += 1
            k += 1
        if pat >= 2 and annotations.symbol[i] == "/" and i > 1 and j < 10:
            marker = annotations.sample[i]
            l = max(0, marker - 90)
            h = l + 180
            test_signal = filt_signal[l:h]
            # downsampling from 360 samples per second to 60 samples per second
            test_signal = test_signal.reshape(30, 6)[:,0]
            test_data[k] = test_signal
            test_labels[k] = 1
            j += 1
            k += 1
```

<h5>Auto-Encoder</h5>
Now, we will build our Auto-Encoder model using Tensorflow and Keras

```python
def AE():
 
    heartbeats = tf.keras.Input(shape=(30,))

    Encode1 = layers.Dense(20, activation = 'relu')(heartbeats)
    Encode2 = layers.Dense(10, activation = 'relu')(Encode1)
    
    Decode1 = layers.Dense(20, activation = 'relu')(Encode2)
    
    reconstruction = layers.Dense(30, activation='tanh')(Decode1)
    
    encoder = tf.keras.Model(heartbeats, Encode2)
    autoencoder = tf.keras.Model(heartbeats, reconstruction)
    
    return encoder, autoencoder
```

We will traing our model for 10,000 epochs with adadelta optimizer and mean-squared loss. The architecture is same as in the paper:  Input-20-10-20-Input'

```python
tf.random.set_seed(42)
encoder, autoencoder = AE()
adadelta = tf.keras.optimizers.Adadelta(learning_rate=0.01)
autoencoder.compile(optimizer = 'adadelta', loss='MSE', metrics=['accuracy'])
hist_ae = autoencoder.fit(training_data, training_data, batch_size=32, initial_epoch=0, epochs=10000, verbose=2, validation_split=0.2, shuffle=True,
                 steps_per_epoch=4)
```

Let us see the lost and accuracy plots for our autoencoder:

```python
plt.plot(hist_ae.history['loss'])
plt.plot(hist_ae.history['val_loss'])
_ = plt.xlabel('Epoch')
_ = plt.ylabel('Mean Squared Error')
```

```python
plt.plot(hist_ae.history['accuracy'])
plt.plot(hist_ae.history['val_accuracy'])
_ = plt.xlabel('Epoch')
_ = plt.ylabel('Accuracy')
```

<h5>$\beta$-Variational Auto Encoder</h5>
In a $\beta$-VAE, we try to learn the latent representation as a distribution, and use the following loss term to train:
$$\mathbb{E}_{q_\phi (z|x)} [\log p_\theta (x|z)] - \beta D_{KL}(q_\phi (z|x)||p(z) )$$

```python
def sampling(z_mu, z_log_var): # reparametrization trick
    batch = tf.shape(z_mu)[0]
    dim = tf.shape(z_mu)[1]
    # by default, tf.random.normal has mu = 0 and sigma = 1.0
    epsilon = tf.random.normal(shape = (batch, dim))
    return z_mu + tf.exp(0.5 * z_log_var) * epsilon
```

```python
heartbeats = tf.keras.Input(shape=(30,))

Encode1 = layers.Dense(20, activation = 'relu')(heartbeats)

z_mean = layers.Dense(10)(Encode1)
z_log_sigma = layers.Dense(10)(Encode1)

z = sampling(z_mean, z_log_sigma)

Decode1 = layers.Dense(10, activation = 'relu')(z)
reconstruction = layers.Dense(30, activation='tanh')(Decode1)

va_encoder = tf.keras.Model(heartbeats, z_mean)
vae = tf.keras.Model(heartbeats, reconstruction)
```

```python
mse_loss = tf.keras.losses.MSE(heartbeats, reconstruction)
kl_loss = - 0.5 * tf.keras.backend.mean(1 + z_log_sigma - tf.keras.backend.square(z_mean) - tf.keras.backend.exp(z_log_sigma), axis=-1)
vae_loss = mse_loss + 5*kl_loss
vae.add_loss(vae_loss)
vae.compile(optimizer = 'adadelta',  metrics=['accuracy'])
hist_vae = vae.fit(training_data, batch_size=32, initial_epoch=0, epochs=10000, verbose=2, 
                   validation_split=0.2, shuffle=True, steps_per_epoch=4)
```

```python
va_embedding = va_encoder.predict(training_data)
va_std_dev = np.std(va_embedding, axis = 0)
bar_width = 0.35
plt.bar(np.arange(10)-bar_width/2, std_dev, width = bar_width, color='blue', label='AE')
plt.bar(np.arange(10)+bar_width/2, va_std_dev, width = bar_width, color='red', label='VAE')
_ = plt.xticks(np.arange(10), np.arange(10))
_ = plt.legend(loc='best')
_ = plt.title("Std. deviation of embedding: Training data")
_ = plt.xlabel("Dimension")
_ = plt.ylabel("Standard deviation")
```
